```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_excel('Data.xlsx')
df.describe()
```

```python
df.plot(x='Year')
plt.show()
```

```python
plt.figure(figsize=(12,6)) 

plt.plot(df['Year'],df['Filing '], label='Filing')
plt.plot(df['Year'],df['Insurance'], label='Insurance')
plt.plot(df['Year'],df['Landscape'], label='Landscape')
plt.plot(df['Year'],df['Legal'], label='Legal')
plt.plot(df['Year'],df['Management'], label='Management')
plt.plot(df['Year'],df['Member'], label='Member')
plt.plot(df['Year'],df['Other'], label='Other')
plt.plot(df['Year'],df['Postage'], label='Postage')

plt.xticks([2022,2023,2024])
plt.xlabel('Year')
plt.ylabel('Cost ($)')
plt.title('Income [ $13,954, $17,256, $17,156 ] //// Expenses [ $15,038, $15,569, $15,636]')
plt.legend()

plt.figure(figsize=(12,6)) 

plt.subplot(1,2,1)

plt.bar(df['Year'],df['Dues'], label='Dues', edgecolor='red', linewidth=2, color='red')
plt.xticks([2022,2023,2024])
plt.xlabel('Year')
plt.ylabel('Cost ($)')
plt.title('Dues')
plt.legend()

plt.subplot(1,2,2)

plt.bar(df['Year'],df['Net'], label='Net', edgecolor='blue', linewidth=2, color='blue')
plt.xticks([2022,2023,2024])
plt.xlabel('Year')
plt.ylabel('Cost ($)')
plt.title('Net')
plt.legend()

plt.show()
```
